```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, roc_curve, average_precision_score, precision_recall_curve, PrecisionRecallDisplay, auc
import matplotlib.pyplot as plt 
from sklearn import metrics 
import matplotlib.pyplot as plt 
from sklearn.model_selection import validation_curve, ValidationCurveDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.model_selection import validation_curve, ValidationCurveDisplay, cross_val_score
from sklearn.ensemble  import GradientBoostingClassifier, ExtraTreesClassifier, RandomForestClassifier
```

```python
df = pd.read_csv('mushrooms.csv')
# df.head()
print(f"{df.shape}")
df = pd.get_dummies(df, columns=['class','cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'], dtype=int)
print(f"{df.shape}")
y = df['class_p']
X = df.drop(columns = ['class_p', 'class_e'])
#Прогнозируем несъедобные грибы
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
#Решающие деревья
cl_tree = DecisionTreeClassifier(random_state=0, max_depth=3)
cl_tree.fit(X_train, y_train)
cl_tree.score(X_test,y_test)
```

# Смотрим cv по максимальной глубине по решающим деревьям

```python
#СV Decision Tree Class
param_range = np.arange(0, 20, 1)
train_scores, test_scores = validation_curve(cl_tree, X, y, param_name="max_depth", param_range=param_range, n_jobs=-1, cv=10, scoring="accuracy")
display = ValidationCurveDisplay( param_name='max_depth', param_range=param_range, train_scores=train_scores, test_scores=test_scores, score_name="accuracy" )
display.plot()
# test_scores
```

```python
#СV Decision Tree Class ROC AUC
param_range = np.arange(0, 10, 1)
train_scores, test_scores = validation_curve(cl_tree, X, y, param_name="max_depth", param_range=param_range, n_jobs=-1, cv=10, scoring="roc_auc")
display = ValidationCurveDisplay( param_name='max_depth', param_range=param_range, train_scores=train_scores, test_scores=test_scores, score_name="roc_auc" )
display.plot()
# test_scores
```

```python
#СV Decision Tree Class ROC AUC
param_range = np.arange(0, 10, 1)
train_scores, test_scores = validation_curve(cl_tree, X, y, param_name="max_leaf_nodes", param_range=param_range, n_jobs=-1, cv=10, scoring="roc_auc")
display = ValidationCurveDisplay( param_name='max_leaf_nodes', param_range=param_range, train_scores=train_scores, test_scores=test_scores, score_name="roc_auc" )
display.plot()
# test_scores
```

Тестовая точность для решающих деревьев выше 95, что в принципе неплохо

# Градиентный бустинг, смотрим показатель точности по количеству деревьев

```python
#Градиентный бустинг

gbr = GradientBoostingClassifier(random_state=0)
gbr.fit(X_train, y_train)
gbr.score(X_test,y_test)

param_range = np.arange(0, 40, 1)
train_scores, test_scores = validation_curve(gbr, X, y, param_name="n_estimators", param_range=param_range, n_jobs=-1, cv=10, scoring="accuracy")
display = ValidationCurveDisplay( param_name='n_estimators', param_range=param_range, train_scores=train_scores, test_scores=test_scores, score_name="accuracy" )
display.plot()
```

```python
#Градиентный бустинг roc_auc

gbr = GradientBoostingClassifier(random_state=0)
gbr.fit(X_train, y_train)
gbr.score(X_test,y_test)

param_range = np.arange(0, 30, 1)
train_scores, test_scores = validation_curve(gbr, X, y, param_name="n_estimators", param_range=param_range, n_jobs=-1, cv=10, scoring="roc_auc")
display = ValidationCurveDisplay( param_name='n_estimators', param_range=param_range, train_scores=train_scores, test_scores=test_scores, score_name="roc_auc" )
display.plot()
```

Не нравится волнообразность в тренировочной и тестовой вариантах, но отлично видно, как модель скорее всего
достигла своего пика на 5 дереве и дальше уже успешно переобучилась
